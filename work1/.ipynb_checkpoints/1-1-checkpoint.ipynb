# 課題1.1

```julia
using Plots
gr(legend=false)

# given cubic formula
p3(x::Float64)::Float64 = (5x^3 - 3x)/2

# p3 diff
p3d(x::Float64)::Float64 = (15x^2 - 3)/2

# true value
const t = √(3/5)
```

## 1.1.1
### 二分法

```julia
#Bisection method error
function bisectionerror(f, tval, a, b)
    # absolute error
    e = []
    for i = 0:20
        # aproximate solution
        c = (a + b)/2
        f(a)*f(c) > 0 && (a = c)
        f(b)*f(c) > 0 && (b = c)
        push!(e, abs(c - tval))
    end
    return e
end

#main
function main()
    ε = bisectionerror(p3, t, 0.1, 0.9)
    println(ε)
    #plot and save
    plot(0:20, ε,
        w=2, marker=1,
        title = "Bisection Method",
        xlabel = "Count",
        ylabel = "Absolute error")
    savefig("img/1_1_1.png")
    plot!()
end


@time main()
```

## 1.1.2
### ニュートン法

$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$

```julia
# Newton method error
function newtonerror(f, fdiff, tval, x0)
    #init x
    x = x0
    # absolute error
    e = []
    for i = 0:20
        # aproximate solution
        x = x - f(x)/fdiff(x)
        push!(e, abs(x - tval))
    end
    return e
end

function main()
    ε = newtonerror(p3, p3d, t, 0.5)
    println(ε)
    #plot and save
    plot(0:20, ε,
        w=2, marker=1,
        title = "Newton Method",
        xlabel = "Count",
        ylabel = "Absolute error")
    savefig("img/1_1_2.png")
    plot!()
end

@time main()
```
