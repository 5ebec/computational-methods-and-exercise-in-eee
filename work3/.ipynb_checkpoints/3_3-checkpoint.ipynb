## 課題3.3
次の微分方程式の初期値問題を4次のルンゲ・クッタ法(古典的ルンゲ・クッタ法)により解くプログラムを作成せよ.これはローレンツ・アトラクタ(Lorenz attractor)と呼ばれるものである.
\begin{eqnarray}
    y'=
    \begin{pmatrix}
        \sigma (y_2 - y_1) \\
        ry_1 - y_2 - y_1y_3 \\
        y_1y_2 - by_3
    \end{pmatrix},
    \quad
    \sigma = 10,\;
    b=8/3,\;
    r=28
\end{eqnarray}

```julia
using OffsetArrays
using LinearAlgebra
using LaTeXStrings
using Plots
pyplot(w=3, legend=false)
ENV["PLOTS_TEST"]="true"

const σ = 10.
const b = 8/3
const r = 28
const Δt = 0.001
const t_max = 10

# diff
diff(y::AbstractArray) = [σ*(y[2] - y[1])
                          r*y[1] - y[2] - y[1]*y[3]
                          y[1]*y[2] - b*y[3]]

# y init
yinit(y0::AbstractArray, n_max::Int) = OffsetArray(hcat(y0, zeros(Float64, 3, n_max)), 1:3, 0:n_max)

# the classical Runge–Kutta method
function RK4!(y::AbstractArray, n_max::Int, Δx::Float64)
    for n = 0:n_max-1
        k₁ = diff(y[:,n])Δx
        k₂ = diff(y[:,n] + k₁/2)Δx
        k₃ = diff(y[:,n] + k₂/2)Δx
        k₄ = diff(y[:,n] + k₃)Δx
        y[:,n+1] = y[:,n] + (k₁+2k₂+2k₃+k₄)/6
    end
    return y
end
```

### 課題3.3.1
時間刻み$\Delta t$を0.001として，$t=0$から$t=10$の時間にわたり数値解($\boldsymbol{y}(t)$とする)を求め，
 - $y_1$の時間発展
 - $y_1$を横軸に，$y_2$を縦軸とした$(y_1(t),y_2(t))$の軌跡
 
を図示せよ．ただし，初期条件を以下のようにする．
\begin{eqnarray}
    \boldsymbol{y}(0) = 
    \begin{pmatrix}
    1 \\ 0 \\20
    \end{pmatrix}
\end{eqnarray}

```julia
# y1 time evolution
function y1timeevo(y0::AbstractArray, n_max::Int, Δx::Float64, filename::String)
    y = yinit(y0, n_max)
    yc = RK4!(y, n_max, Δx)
    plot!((0:n_max)*Δx, yc[1,0:n_max], xlabel=L"t", ylabel=L"y_1")
    savefig("img/$(filename).png")
    return
end

plot(title=L"y_1(t) \; (y(0)=[1; 0; 20])")
y1timeevo([1.; 0.; 20.], Int(t_max/Δt), Δt, "3_3_1-a")
```

```julia
# (y1, y2) locus
function y1y2locus(y0::AbstractArray, n_max::Int, Δx::Float64, filename::String)
    y = yinit(y0, n_max)
    yc = RK4!(y, n_max, Δx)
    plot!(yc[1,0:n_max], yc[2,0:n_max], xlabel=L"y_1", ylabel=L"y_2")
    savefig("img/$(filename).png")
    return
end

plot(title=L"(y_1(t),y_2(t)) \; (y(0)=[1; 0; 20])")
y1y2locus([1.; 0.; 20.], Int(t_max/Δt), Δt, "3_3_1-b")
```

### 課題3.3.2


```julia
# main
function main()
    for q = [1e-2, 1e-3]
        plot(title=L"y_1(t) \; (y(0)=[1+"*string(q)*L"; 0; 20])")
        y1timeevo([1+q; 0.; 20.], Int(t_max/Δt), Δt, "3_3_2-a-$(q)")
        plot(title=L"(y_1(t),y_2(t)) \; (y(0)=[1+"*string(q)*L"; 0; 20])")
        y1y2locus([1+q; 0.; 20.], Int(t_max/Δt), Δt, "3_3_2-b-$(q)")
    end
end

main()
```

## 初期値鋭敏性

```julia
# plot
function locus(method!::Function, y0::AbstractArray, n_max::Int, Δx::Float64)
    y = yinit(y0, n_max)
    yc = method!(y, n_max, Δx)
    plot3d(yc[1,0:n_max], yc[2,0:n_max], yc[3,0:n_max], 
        xaxis=("x", (-25,25)), yaxis=("y", (-25,25)), zaxis=("z", (0,50)),
        title = "Lorenz Attractor", w=2)
    return
end

anim = @animate for q = 0:1e-2:1
    locus(RK4!, [1+q; 0.; 20.], Int(t_max/Δt), Δt)
end

gif(anim, "img/3_3-anim.gif", fps=30)
```
