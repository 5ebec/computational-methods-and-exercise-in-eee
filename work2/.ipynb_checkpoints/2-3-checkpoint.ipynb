## 課題2.3(一部必須)
二次元空間中での電界$E$の計算方法を考える。計算対象とする空間内に電荷が存在しない場合
$$\nabla \cdot E = 0 \tag{3-1}$$
がGaussの法則より成立するため，電位$E=-\nabla \phi$を導入すると，
$$\nabla^2\phi=0\tag{3-2}$$
となる．この式はLaplace方程式と呼ばれる．即ち，二次元空間内においては，
$$\frac{\partial^2\phi}{\partial x^2}+\frac{\partial^2\phi}{\partial y^2}=0\tag{3-3}$$
が成立する．

### 2.3.1(必須)
式(3-3)を、中心差分式を示せ。ただし、ここでは簡単のため、$\Delta x = \Delta y = h$としてよい。



```julia
function laplaceeq(ϕ::Array{Float64, 2}, h::Int)
    (circshift(ϕ, (0, h)) + circshift(ϕ, (0, -h)) +
     circshift(ϕ, (h, 0)) + circshift(ϕ, (-h, 0)) - 4ϕ) / h^2
end
```

以下の設問では、図2.3.1に示すような二つの導体の間の空間の電位をLaplace方程式を解くことによって求めることを考える。下側の導体は接地され電位0V、上側の導体は電圧源に接続され電位100Vに保たれているものとする。$g$は$x = 0$周辺を除いたギャップの大きさ、$d, w$は$x = 0$周辺に存在する下部導体の凹部の寸法である。また、$l$は格子分割を行う$x$方向の範囲であり、可変である。

### 2.3.2(必須)
$g = 10 \,\mathrm{mm}, w = 10\,\mathrm{mm}, d = 10\,\mathrm{mm}, l = 30 \,\mathrm{mm}$とし、図2.3.2のように$\Delta x = \Delta y = h = 5 \mathrm{mm}$で格子分割する。 境界条件として、$\phi_5 = 50 \,\mathrm{V}$ という固定境界条件を設定する。電位$\phi_0, \phi_1, ..., \phi_4$を解とする連立方程式をたてて、手計算で解いてみよ。
![図2.3.2](fig2-3-2.png)

```julia
using SymPy

@syms ϕ0 ϕ1 ϕ2 ϕ3 ϕ4

f1 = 4ϕ0 - ϕ1 ⩵ 0
f2 = ϕ0 - 4ϕ1 + ϕ2 ⩵ 0
f3 = ϕ1 - 4ϕ2 + 2ϕ3 ⩵ -100
f4 = ϕ2 - 4ϕ3 + ϕ4 ⩵ -100
f5 = ϕ3 - 4ϕ4 ⩵ -150

s = solve([f1, f2, f3, f4, f5])
println(s)
```

### 2.3.3(必須)
図2.3.1に示す解析対象について、任意の$g, w, d, l, h$に対して、格子点のポテンシャルを求めるプログラムを作成せよ。2.1.4で作った任意の連立線形方程式を解くサブルーチン、もしくは、2.2.1で作った任意の連立線形方程式を解くサブルーチンを使うこと。  
ただし、$x = −l / 2, l / 2$における境界条件として、位置$y$における電位$\phi$が以下の関数で与えられる固定境界条件を用いるものとする。
$$\phi =\frac{y}{g}\times 100\,\mathrm{V} \tag{3-4}$$
2.3.2の問題をこのプログラムにより解き，正しい解が得られることを確認せよ．

```julia
# load custom module "LUmodule"
push!(LOAD_PATH, "./LUmodule/src")
using LUmodule


# Lattice point's potential
function potential(g, w, d, l, h)
    
    # boundary
    y0, y1, y2 = 1, Int(g/h+1), Int((g+d)/h+1)
    x0, x1, x2, x3 = 1, Int((l-w)/2h+1), Int((l+w)/2h+1), Int(l/h+1)
    
    # set ϕ
    ϕ = ones(y2, x3)
    ϕ[y0, :] .= 100.
    ϕ[y2, :] .= 0.
    ϕ[y1:y2, x0:x1] .= ϕ[y1:y2, x2:x3] .= 0.
    for y = y0+1:y1-1
        ϕ[y, x0] = ϕ[y, x3] = 100(y1-y)h/g
    end
    
    # set A, b
    n = y2*x3
    A = one(zeros(n, n))
    b = zeros(n)
    for i=1:n
        if ϕ[i] == 1.
            A[i, i-1] = A[i, i+1] = A[i, i-y2] = A[i, i+y2] = -1/4
        else
            b[i] = ϕ[i]
        end
    end
    
    # LU solve
    ϕ = reshape(lusolve!(A, b), (y2, x3))
    ϕ[y1+1:y2, x0:x1-1] .= ϕ[y1+1:y2, x2+1:x3] .= NaN
    
    return ϕ
end

potential(10, 10, 10, 30, 5)
```

### 2.3.4
2.3.3で作成したプログラムを用いて、$g = 10 \,\mathrm{mm}, w = 10 \,\mathrm{mm}, d =10 \,\mathrm{mm}, l = 10, 30, 50 \,\mathrm{mm}, h = 5 \,\mathrm{mm} $として電位分布を計算し、適当な形でグラフ化せよ(Excel等でも可)。$l$が計算結果に与える影響について調べ、その理由について考察せよ。

```julia
using Plots
gr()
for l = [10 30 50]
    @time heatmap(potential(10, 10, 10, l, 5), 
        yaxis=:flip, 
        fill=true, 
        c=:viridis, 
        title="g=10mm,w=10mm,d=10mm,l="*string(l)*"mm,h=5mm",
        aspect_ratio=1)
    savefig("2_3_4-l="*string(l)*".pdf")
end
plot!()
```

### 2.3.5
2.3.3で作成したプログラムを用いて、$g = 10 \,\mathrm{mm}, w = 10 \,\mathrm{mm}, d = 10 \,\mathrm{mm}, l = 30 \,\mathrm{mm}, h = 1, 2, 5 \,\mathrm{mm}$として電位分布を計算し、適当な形でグラフ化せよ(Excel等でも可)。$h$が計算結果に与える影響について調べ よ。

```julia
using Plots
gr()
for h = [5 2 1]
    @time heatmap(potential(10, 10, 10, 30, h), 
        yaxis=:flip, 
        fill=true, 
        c=:viridis, 
        title="g=10mm,w=10mm,d=10mm,l=30mm,h="*string(h)*"mm",
        aspect_ratio=1)
    savefig("2_3_5-h="*string(h)*".pdf")
end
plot!()
```
