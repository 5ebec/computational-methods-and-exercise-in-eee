## 課題2.2(一部必須)
$n$元連立一次方程式
$$
A\boldsymbol{x}=\boldsymbol{b}
\tag{2-1}
$$
を反復法で解く場合，式(2-1)を
$$
\boldsymbol{x}=g(\boldsymbol{x})=B\boldsymbol{x}+\boldsymbol{c}
\tag{2-2}
$$
と同値変形し，次式のように$\boldsymbol{x}^{(0)}$から出発して，逐次代入$\boldsymbol{x}^{(k+1)}=g\left(\boldsymbol{x}^{(k)}\right)$を行って解を求める．
$$
\boldsymbol{x}^{(k+1)}=B\boldsymbol{x}^{(k)}+\boldsymbol{c}
\tag{2-3}
$$
今，行列$A$を下記のように$M,N$に分解する．
$$
A=M-N
\tag{2-4}
$$
これにより，式(2-1)は次式のように同値変形できる．
$$
M\boldsymbol{x}=N\boldsymbol{x}+\boldsymbol{b}
\tag{2-5}
$$
この式から，式(2-3)に相当するものとして，次のような反復式(逐次代入式)を構成できる．
$$
M\boldsymbol{x}^{(k+1)}=N\boldsymbol{x}^{(k)}+\boldsymbol{b}
\tag{2-6}
$$
$$
\boldsymbol{x}^{(k+1)}=M^{-1}N\boldsymbol{x}^{(k)}+M^{-1}b
\tag{2-7}
$$
今，任意の行列$A$は，以下に示すように対角行列$D$、下三角行列$L'$、上三角行列$U'$の和として表記できる。
$$
D=\begin{bmatrix}
a_{11} & 0 & \cdots & 0 \\
0 & a_{22} & \ddots & \vdots \\
\vdots & \ddots & \ddots & 0 \\
0 & \cdots & 0 & a_{nn}
\end{bmatrix},
L'=\begin{bmatrix}
0 & 0 & \cdots & 0 \\
a_{21} & \ddots & \ddots & \vdots \\
\vdots & \ddots & \ddots & 0 \\
a_{n1} & \cdots & a_{n,n-1} & 0
\end{bmatrix},
U'=\begin{bmatrix}
0 & a_{12} & \cdots & a_{1n} \\
0 & \ddots & \ddots & \vdots \\
\vdots & \ddots & \ddots & a_{n-1,n} \\
0 & \cdots & 0 & 0 
\end{bmatrix}
\tag{2-8}
$$
$$A=D+L'+U'\tag{2-9}$$
このことを利用し，式(2-4)のように$A$を分解するとき，$M=D,N=-(L'+U')$として構成した反復法をヤコビ法と呼ぶ．即ち，式(2-6),(2-7)より，ヤコビ法の反復式は以下のようになる．
$$D\boldsymbol{x}^{(k+1)}=-(L'+U')\boldsymbol{x}^{(k)}+\boldsymbol{b}\tag{2-10}$$
$$\boldsymbol{x}^{(k+1)}=-D^{-1}(L'+U')\boldsymbol{x}^{(k)}+D^{-1}\boldsymbol{b}\tag{2-11}$$
式(2-10)の右辺第1項を移項して成分表示すると以下のようになる．
$$
a_{11}x_1^{(k+1)}+a_{12}x_2^{(k)}+\cdots+a_{1n}x_n^{(k)}=b_1\\
a_{21}x_1^{(k)}+a_{22}x_2^{(k+1)}+\cdots+a_{2n}x_n^{(k)}=b_2\\
\vdots+\vdots+\cdots+\vdots=\cdots \\
a_{n1}x_1^{(k)}+a_{n2}x_2^{(k)}+\cdots+a_{nn}x_n^{(k+1)}=b_n
\tag{2-12}
$$
この$n$本の式のうち$i$番目の式により，$x_i^{(k)}(l=1,\ldots,n(\mathrm{excluding} \ i))$から$x_i^{(k+1)}$を求めることができる．

### 2.2.1(必須)
式(2-12)により、$x^{(k)}$から$x^{(k+1)}$を計算するサブルーチンを作成せよ。さらにこれを使って、ヤコビ法により任意の連立線形方程式を解くサブルーチンを作成せよ。このサブルーチンを動作させるメインプログラムを作成し、適当な問題を考え、正しい解が得られることを確認せよ。


```julia
# x's recurrence relation
function xrecur(A, x0, b)
    n = size(b)[1]
    x1 = zeros(n)
    for i = 1:n
        x1[i] = b[i]
        for j = 1:n
            x1[i] -= (j != i ? A[i,j]*x0[j] : 0.)
        end
        x1[i] /= A[i,i]
    end
    return x1
end


# Jacobi method
function jacobi(A, x0, b)
    #init
    x1 = x0
    for i=1:150
        x1 = xrecur(A, x0, b)
        if sum(abs, x1 - x0) < 1e-9
            break
        else
            x0 = x1
        end
    end
    return x1
end


# main
function main()   
    A = [3. 1. 1.
         1. 3. 1. 
         1. 1. 3.]
    b = [0.; 4.; 6.]
    x0 = zeros(size(b)[1])

    println("A=", A)
    println("b=", b)
    println("x=", jacobi(A, x0, b))
end


main()
```

### 2.2.2
前問のサブルーチンを使って、任意の$n$に対応して、式(1-6)を解くプログラムを作成せよ。
手計算のできる範囲の小さな$n$(ただし、$n \geq 3$)で動作確認をするとともに、大きな$n$について式(1-6)を解いてみよ。ただし,$a$の値が、$a = −0.25, −0.75$の場合について解け。また、初期ベクトル$x^{(0)}$の与え方について検討せよ。

```julia
using LinearAlgebra
using Plots
gr(ylabel="element", ms=2, legend=false)


# x calculation
function xrecur(A, x0, b)
    n = size(b)[1]
    x1 = zeros(n)
    for i = 1:n
        x1[i] = b[i]
        for j = 1:n
            x1[i] -= (j != i ? A[i,j]*x0[j] : 0.)
        end
        x1[i] /= A[i,i]
    end
    return x1
end


# Jacobi method
function jacobi(A, x0, b, filename)
    #init
    x1 = x0
    anim = @animate for i=1:150
        x1 = xrecur(A, x0, b)
        if sum(abs, x1 - x0) < 1e-7
            break
        else
            x0 = x1
        end
        scatter(x1, title=filename)
    end
    gif(anim, "2-2_"*filename*".gif", fps = 10)
    return x1
end


# A matrix
function amtx(n, a)
    mtx = Array(SymTridiagonal(fill(1., n), fill(a, n-1)))
    mtx[1,n] = mtx[n,1] = a
    return mtx
end


# B vector
function bvec(n)
    vec = zeros(n)
    vec[1] = 1
    vec[n] = -1
    return vec
end


# main
function main()
    n = 50
    A1 = amtx(n, -1/4)
    A2 = amtx(n, -3/4)
    b = bvec(n)
    
    println("x=", jacobi(A1, zeros(n), b, "x0=zeros a=-0.25"))
    println("x=", jacobi(A2, zeros(n), b, "x0=zeros a=-0.75"))
    println("x=", jacobi(A1, 1e3*rand(n), b, "x0=rand a=-0.25"))
    println("x=", jacobi(A2, 1e2*rand(n), b, "x0=rand a=-0.75"))
end

main()
```

### 2.2.3
式(1-6)をLU分解とヤコビ法によって求解する際に要する計算時間を、横軸$n$、縦軸を計算時間として図示せよ。計算時間の測り方については各自で調べ工夫せよ。

```julia
using LinearAlgebra
using Plots
gr(title = "LU decomposition vs. Jacobi method",
    xlabel = "n",
    ylabel = "calculation time")


# LU decomposition
function ludecomp!(A)
    n = size(A)[1]
    for j = 1:n
        # u_{ij} (i <= j)
        for i = 1:j
            u = A[i,j]
            for k = 1:(i-1)
                u -= A[i,k]*A[k,j]
            end
            A[i,j] = u
        end
        # l_{ij} (i > j)
        for i = (j+1):n
            l = A[i,j]
            for k = 1:(j-1)
                l -= A[i,k]*A[k,j]
            end
            A[i,j] = l/A[j,j]
        end
    end
    return A
end


# Forward Substitution
function fwdsub!(L, b)
    for i = 1:size(b)[1]
        y = b[i]
        for k = 1:i-1
            y -= b[k]*L[i,k]
        end
        b[i] = y
    end
    return b
end


# Backward Substitution
function bwdsub!(U, y)
    n = size(y)[1]
    for i = n:-1:1
        x = y[i]
        for k = (i+1):n
            x -= y[k]*U[i,k]
        end
        y[i] = x/U[i,i]
    end
    return y
end


# simultaneous linear equations solution
function lusolve!(A, b)
    ludecomp!(A)
    fwdsub!(A, b)
    bwdsub!(A, b)
end


# x calculation
function xrecur(A, x0, b)
    n = size(b)[1]
    x1 = zeros(n)
    for i = 1:n
        x1[i] = b[i]
        for j = 1:n
            x1[i] -= (j != i ? A[i,j]*x0[j] : 0.)
        end
        x1[i] /= A[i,i]
    end
    return x1
end


# Jacobi method
function jacobi(A, x0, b)
    #init
    x1 = x0
    for i=1:150
        x1 = xrecur(A, x0, b)
        if sum(abs, x1 - x0) < 1e-10
            break
        else
            x0 = x1
        end
    end
    return x1
end


# A matrix
function amtx(n, a)
    mtx = Array(SymTridiagonal(fill(1., n), fill(a, n-1)))
    mtx[1,n] = mtx[n,1] = a
    return mtx
end


# b vector
function bvec(n)
    vec = zeros(n,1)
    vec[1] = 1
    vec[n] = -1
    return vec
end


# main
function main()
    maxn = 225
    jacobit = []
    lut = []
    anim = @animate for n = 1:maxn
        A1 = amtx(n, -1/4)
        b = bvec(n)
        push!(jacobit, @elapsed jacobi(A1, rand(n), b))
        push!(lut, @elapsed lusolve!(A1, b))
        plot(jacobit, lab="Jacobi")
        plot!(lut, lab="LU")
        xlims!(0, maxn)
    end
    gif(anim, "2-3_1.gif", fps = 20)
    savefig("2-3_1.pdf")
    plot!()
end


main()
```
